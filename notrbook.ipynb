```python
df["id"]=df["id"].astype(int)
```

```python
df
```

```python
d={}
for i in range(len(df)):
    if df["id"][i] in d:
        (d[df["id"][i]])[df["item_name"][i]] =1
            
    else:
        d[df["id"][i]]={df["item_name"][i]:1}
```

```python
d
```

```python
li=[]
for i in d:
    for j in d[i]:
        li.append([i,j,d[i][j]])
data = pd.DataFrame(li,columns=["id","item_name","purchase_count"])
```

```python
data
```

```python
data=data.groupby('id').filter(lambda x : len(x)>3)
```
